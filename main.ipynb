##### 1. TELEMETRIA

```python
import random

def obter_dados_sistema():
    """
    Simula uma chamada para uma API que retorna dados da nave
    """
    dados = {
        "temperatura_interna": random.uniform(15, 30),   # °C
        "temperatura_externa": random.uniform(-20, 45),  # °C
        "integridade_estrutural": random.choice([0, 1]), # 1 = ok
        "energia_capacidade_total": 1000,                # kWh
        "carga_atual_percentual": random.uniform(50, 100),
        "consumo_decolagem": 400,                        # kWh
        "perdas_energeticas": random.uniform(10, 50),    # kWh
        "pressao_tanques": random.uniform(2.5, 5.0),     # bar
        "modulos_criticos": {
            "navegacao": random.choice([True, False]),
            "comunicacao": random.choice([True, False]),
            "controle_voo": random.choice([True, False])
        }
    }
    
    return dados
```

##### 2. VERIFICAÇÃO DAS CONDIÇÕES

```python
def verificar_condicoes(dados):

    # Faixas seguras
    TEMP_INTERNA_MIN = 18
    TEMP_INTERNA_MAX = 27

    TEMP_EXTERNA_MIN = -10
    TEMP_EXTERNA_MAX = 40

    PRESSAO_MIN = 3.0
    PRESSAO_MAX = 4.5

    # 1 - Verificar temperaturas
    if not (TEMP_INTERNA_MIN <= dados["temperatura_interna"] <= TEMP_INTERNA_MAX):
        return "DECOLAGEM ABORTADA - Temperatura interna fora da faixa segura"

    if not (TEMP_EXTERNA_MIN <= dados["temperatura_externa"] <= TEMP_EXTERNA_MAX):
        return "DECOLAGEM ABORTADA - Temperatura externa fora da faixa segura"

    # 2 - Integridade estrutural
    if dados["integridade_estrutural"] != 1:
        return "DECOLAGEM ABORTADA - Falha estrutural"

    # 3 - Pressão dos tanques
    if not (PRESSAO_MIN <= dados["pressao_tanques"] <= PRESSAO_MAX):
        return "DECOLAGEM ABORTADA - Pressão dos tanques inadequada"

    # 4 - Módulos críticos
    if not all(dados["modulos_criticos"].values()):
        return "DECOLAGEM ABORTADA - Módulos críticos indisponíveis"

    # 5 - Energia disponível
    capacidade = dados["energia_capacidade_total"]
    carga = dados["carga_atual_percentual"] / 100

    energia_disponivel = capacidade * carga
    energia_necessaria = dados["consumo_decolagem"] + dados["perdas_energeticas"]

    if energia_disponivel < energia_necessaria:
        return "DECOLAGEM ABORTADA - Energia insuficiente"

    return "PRONTO PARA DECOLAR"
```

##### 3. ANÁLISE IA

```python
def analise_ia(dados):

    classificacao = {}
    anomalias = []
    riscos = []

    # Classificação simples dos parâmetros
    if dados["temperatura_interna"] > 25:
        classificacao["temperatura_interna"] = "ALTA"
    else:
        classificacao["temperatura_interna"] = "NORMAL"

    if dados["pressao_tanques"] > 4.3:
        classificacao["pressao_tanques"] = "PRÓXIMA DO LIMITE"

    # Detecção simples de anomalias
    if dados["temperatura_externa"] < -5:
        anomalias.append("Temperatura externa muito baixa")

    if dados["pressao_tanques"] > 4.4:
        anomalias.append("Pressão próxima do limite máximo")

    # Sugestão de riscos
    if dados["carga_atual_percentual"] < 30:
        riscos.append("Margem energética baixa para operações futuras")

    if not all(dados["modulos_criticos"].values()):
        riscos.append("Possível falha de subsistema crítico")

    return classificacao, anomalias, riscos
```

##### 4. EXECUÇÃO

```python
def main():

    dados = obter_dados_sistema()

    print("\n================ TELEMETRIA RECEBIDA ================\n")

    print("Temperaturas")
    print(f"  Interna : {dados['temperatura_interna']:.2f} °C")
    print(f"  Externa : {dados['temperatura_externa']:.2f} °C")

    print("\nEstrutura")
    status_estrutura = "OK" if dados["integridade_estrutural"] == 1 else "FALHA"
    print(f"  Integridade estrutural : {status_estrutura}")

    print("\nEnergia")
    print(f"  Capacidade total       : {dados['energia_capacidade_total']} kWh")
    print(f"  Carga atual            : {dados['carga_atual_percentual']:.2f} %")
    print(f"  Consumo decolagem      : {dados['consumo_decolagem']} kWh")
    print(f"  Perdas energéticas     : {dados['perdas_energeticas']:.2f} kWh")

    print("\nPropulsão")
    print(f"  Pressão dos tanques    : {dados['pressao_tanques']:.2f} bar")

    print("\nMódulos críticos")
    for modulo, status in dados["modulos_criticos"].items():
        estado = "OK" if status else "FALHA"
        print(f"  {modulo:15} : {estado}")

    resultado = verificar_condicoes(dados)

    print("\n================ RESULTADO DA VERIFICAÇÃO ================")
    print(resultado)

    classificacao, anomalias, riscos = analise_ia(dados)

    print("\n================ ANÁLISE ASSISTIDA POR IA ================\n")

    print("Classificação dos dados")
    for k, v in classificacao.items():
        print(f"  {k:20} : {v}")

    print("\nAnomalias detectadas")
    if anomalias:
        for a in anomalias:
            print(f"  - {a}")
    else:
        print("  Nenhuma anomalia detectada")

    print("\nSugestões de risco")
    if riscos:
        for r in riscos:
            print(f"  - {r}")
    else:
        print("  Nenhum risco relevante identificado")

    print("\n==========================================================\n")


if __name__ == "__main__":
    main()
```
